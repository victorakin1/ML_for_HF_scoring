```python
import pandas as pd
import os
from pyAudioAnalysis import audioTrainTest as aT

def prediction(file_sprdsheet, test_files):
    
    SVM_val_df = pd.read_excel(file_sprdsheet)
    
    SVM_val_df["pred_1"] = 0
    SVM_val_df["pred_2"] = 0
    SVM_val_df["pred_3"] = 0  
    SVM_val_df["pred_4"] = 0  
    #SVM_val_df["pred_5"] = 0 
    par_path = "C:/Users/akingv01/OneDrive - BBC/ML_AM_Monitoring"
    
    source_files = os.path.join(par_path, test_files)
    filename_list = os.listdir(source_files)
    
    for i in filename_list:
        for j in SVM_val_df.index:
            
            if SVM_val_df["Filename"][j] in i:
                
                #print(True)
                dir_file = os.path.join(source_files,i)
                c, p, p_nam = aT.file_classification(dir_file, "svm_1-5","svm_rbf")
                SVM_val_df.at[j, "pred_1"] = p[0]
                SVM_val_df.at[j, "pred_2"] = p[1]
                SVM_val_df.at[j, "pred_3"] = p[2]
                SVM_val_df.at[j, "pred_4"] = p[3]
                #SVM_val_df.at[j, "pred_5"] = p[4]
    SVM_val_df["SVM_predict"] = 0
    for i in SVM_val_df.index:
        pred_list = [SVM_val_df["pred_1"][i],SVM_val_df["pred_2"][i], SVM_val_df["pred_3"][i], SVM_val_df["pred_4"][i]]
        pred_val = ['1','2','3','4']
        pred_list.index(max(pred_list))
        SVM_val_df.at[i, "SVM_predict"] = pred_val[pred_list.index(max(pred_list))]
        
    SVM_val_df["dev"] = 0
    for i in SVM_val_df.index:
        SVM_val_df.at[i, "dev"] = int(SVM_val_df["O"][i]) - int(SVM_val_df["SVM_predict"][i])
    svm_val = SVM_val_df["pred_1"] != 0 
    svm_val.to_excel("prediction " + file_sprdsheet.replace(".xlsx","") + ".xlsx", index = False)
    
    return svm_val
                
    
```

```python
import os 
import io
from google.cloud import speech
import pandas as pd

def confidence_score():
    SVM_val_df = pd.read_excel("Data result compare 4 classify.xlsx")
    source_files = "C:/Users/akingv01/OneDrive - BBC/ML_AM_Monitoring/WAV_file_source4 AFES_files"
    filename_list = os.listdir(source_files)
    
    SVM_val_df["confidence"] = 0
    
    for i in filename_list:
        for j in SVM_val_df.index:
            
            if SVM_val_df["Filename"][j] in i:
                
                print(True)
                dir_file = os.path.join(source_files,i)
                
                client = speech.SpeechClient()

                speech_file = dir_file

                with open(speech_file, "rb") as audio_file:
                    content = audio_file.read()

                audio = speech.RecognitionAudio(content=content)

                config = speech.RecognitionConfig(
                    encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
                    sample_rate_hertz=44100,
                    language_code="en-US",
                    enable_word_confidence=True,
                )
                
                response = client.recognize(config=config, audio=audio)
                result = response.results
                
                if len(result) != 0:
                    res = result[0]
                    alternative = res.alternatives[0]
                    print(alternative.confidence)
                    SVM_val_df.at[j, "confidence"] = alternative.confidence
                
    SVM_val_df.to_excel("Data result with confidence v2.xlsx", index = False)


    return SVM_val_df
```

```python
import os 
import pandas as pd 
from pyAudioAnalysis import audioSegmentation as aS
from pyAudioAnalysis.audioSegmentation import mid_term_file_classification, labels_to_segments
from pyAudioAnalysis.audioTrainTest import load_model

path = "C:/Users/akingv01/OneDrive - BBC/ML_AM_Monitoring/AM_Monitoring_Classes/Sorted O score folders_AFES"

def segment_noise_check():
    sgmt_var_df = pd.DataFrame()
    
    sgmt_dev = []
    filename = []
    
    
    for i in os.listdir(path):
        o_scr_pth = os.path.join(path, i)
        #print(o_scr_pth)
        for j in os.listdir(o_scr_pth):
            #print(os.path.join(o_scr_pth,j))
            
            labels, class_names, _, _ = mid_term_file_classification( os.path.join(o_scr_pth,j), 
                                                         "svm_1-5", "svm_rbf",  True, 
                                                         "test_3.segments")
            segs, c = labels_to_segments(labels, mt_step)
            dev = int(class_names[int(max(c))]) - int(class_names[int(min(c))])
            filename.append(j)
            sgmt_dev.append(dev)
    sgmt_var_df.insert(0,"Filename",filename,True)
    sgmt_var_df.insert(1,"Segment Deviation",sgmt_dev,True)
    sgmt_var_df.to_excel("Segment Deviation data.xlsx", index = False)
    
    
    return sgmt_var_df
            
            
    

segment_noise_check()
```

```python
def sum_odd_and_even(lst):
    odd_sum  = 0
    even_sum = 0 
    for i in lst:
        if i%2 == 0:
            even_sum += i
        else:
            odd_sum += i

    return [even_sum,odd_sum]
```
